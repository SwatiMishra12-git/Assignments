```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)  # For reproducibility
study_hours = np.random.uniform(1, 10, 100)  # Random study hours between 1 and 10
exam_scores = 5 * study_hours + np.random.normal(0, 5, 100)  # Linear relation with noise

import pandas as pd
df = pd.DataFrame({"Study Hours": study_hours, "Exam Score": exam_scores})

plt.figure(figsize=(8, 6))
sns.scatterplot(x="Study Hours", y="Exam Score", data=df, color="blue", alpha=0.7)

plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.title("Scatter Plot of Study Hours vs. Exam Score")

plt.grid(True)
plt.show()
```

**QUESTION 5**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)

# Generate correlated data
size = 100  # Number of samples
X1 = np.random.normal(loc=50, scale=10, size=size)  # Base feature
X2 = X1 * 0.8 + np.random.normal(scale=5, size=size)  # Correlated with X1
X3 = X1 * -0.6 + np.random.normal(scale=5, size=size)  # Negatively correlated with X1
X4 = np.random.normal(loc=30, scale=10, size=size)  # Independent feature

# Create DataFrame
data = pd.DataFrame({'Feature_1': X1, 'Feature_2': X2, 'Feature_3': X3, 'Feature_4': X4})

# Compute Correlation Matrix
correlation_matrix = data.corr()

# Set figure size
plt.figure(figsize=(8, 6))

# Create a Heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Title
plt.title("Correlation Matrix Heatmap", fontsize=14)

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import plotly.express as px

np.random.seed(30)

# Generate random data for X, Y, Z coordinates
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}

df = pd.DataFrame(data)

fig = px.scatter_3d(df, x='X', y='Y', z='Z', title="3D Scatter Plot",
                     labels={'X': 'X-Axis', 'Y': 'Y-Axis', 'Z': 'Z-Axis'},
                     color=df['Z'])

fig.show()
```

```python
import numpy as np
import pandas as pd
import plotly.express as px

np.random.seed(20)

data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

df = pd.DataFrame(data)

fig = px.density_heatmap(df, x='Month', y='Day', z='Sales',
                         color_continuous_scale='Viridis',
                         title="Sales Variation Across Months and Days",
                         labels={'Month': 'Month', 'Day': 'Day', 'Sales': 'Sales'})

fig.show()
```

**QUESTION 4**

```python
import numpy as np
import pandas as pd
import plotly.express as px

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

x, y = np.meshgrid(x, y)

z = np.sin(np.sqrt(x**2 + y**2))

data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

fig = px.scatter_3d(df, x='X', y='Y', z='Z', color='Z',
                     title="3D Surface Plot of z = sin(√(x² + y²))",
                     labels={'X': 'X-Axis', 'Y': 'Y-Axis', 'Z': 'Z-Value'})

fig.show()
```

**QUESTION 5**

```python
from bokeh.plotting import figure, show
from bokeh.models import LinearColorMapper, ColorBar, ColumnDataSource
from bokeh.transform import linear_cmap
import numpy as np
from bokeh.io import output_notebook
output_notebook()


data_heatmap = np.random.rand(10, 10)
x = np.arange(10)
y = np.arange(10)
xx, yy = np.meshgrid(x, y)

x_flat = xx.flatten()
y_flat = yy.flatten()
values = data_heatmap.flatten()

source = ColumnDataSource(data=dict(x=x_flat, y=y_flat, values=values))

mapper = LinearColorMapper(palette="Viridis256", low=min(values), high=max(values))

p = figure(title="Heatmap", x_axis_label="X", y_axis_label="Y", x_range=(-0.5, 9.5), y_range=(-0.5, 9.5),
           toolbar_location=None, tools="")

p.rect(x="x", y="y", width=1, height=1, source=source,
       fill_color=linear_cmap("values", "Viridis256", min(values), max(values)),
       line_color=None)

color_bar = ColorBar(color_mapper=mapper, location=(0, 0))
p.add_layout(color_bar, "right")

show(p)
```
